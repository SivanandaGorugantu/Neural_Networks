```python
import os
# os.environ['CUDA_VISIBLE_DEVICES'] = '0'
import tensorflow as tf
import keras
import keras.backend as K
print(tf.__version__)
```

```python
equation = lambda a,b,c,x: a*(x**2) + b*x + c


model = keras.models.Sequential()

model.add(keras.layers.InputLayer((4,)))
model.add(keras.layers.Dense(16,activation='relu'))
model.add(keras.layers.Dense(32,activation='relu'))
model.add(keras.layers.Dense(64,activation='relu'))
model.add(keras.layers.Dense(128,activation='relu'))
# model.add(keras.layers.Dense(16,))
model.add(keras.layers.Dense(1,activation='relu'))

model.summary()
```

```python
import numpy as np

inpX = []
Y = []
for i in range(100000):
    a = np.random.randint(10)
    b = np.random.randint(10)
    c = np.random.randint(10)
    x = np.random.randint(10)
    
    y = equation(a,b,c,x)
    
    inpX.append([a,b,c,x])
    Y.append(y)
    
# print(inpX,Y)

inpX = np.array(inpX)
Y = np.array(Y)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(inpX, Y, test_size=0.1)
print(X_train.shape,Y_train.shape)
```

```python
model.compile(optimizer=keras.optimizers.Adam(),loss="mse",metrics=['accuracy'])
```

```python
model.fit(X_train,Y_train,batch_size=16,epochs=10,verbose=1,shuffle=True,initial_epoch=0)
```

```python
model.fit(X_train,Y_train,batch_size=64,epochs=50,verbose=1,shuffle=True,initial_epoch=0)
```

```python
model.fit(X_train,Y_train,batch_size=256,epochs=100,verbose=1,shuffle=True,initial_epoch=0)
```

```python
model.compile(optimizer=keras.optimizers.Adam(lr=1e-4),loss="mse",metrics=['accuracy'])
model.fit(X_train,Y_train,batch_size=256,epochs=100,verbose=1,shuffle=True,initial_epoch=0)
```

```python
model.compile(optimizer=keras.optimizers.Adam(lr=1e-5),loss="mse",metrics=['accuracy'])
model.fit(X_train,Y_train,batch_size=256,epochs=100,verbose=1,shuffle=True,initial_epoch=0)
```

```python
y_pred = model.predict(X_test)
```

```python
print(np.squeeze(y_pred[:10]),np.array(Y_test,dtype=np.float32)[:10])
print(type(np.squeeze(y_pred)[0]),type(np.array(Y_test,dtype=np.float32)[0]))
```

```python
model.evaluate(X_test,Y_test)
```
